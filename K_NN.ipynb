```python
# This script uses K-NN classification to find out if based upon their Age and Estimated Salary,
# can we predict if a customer buys a product or not.
# The dataset used here is 'Social_Network_Ads'.
```

```python
# 1. Data Pre-processing
```

```python
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import the dataset
dataset = pd.read_csv('Social_Network_Ads.csv')

# set the independent variables
X = dataset.iloc[:,:-1].values

# set the dependent variable
Y = dataset.iloc[:,-1].values
```

```python
# 2. Split dataset into training and testing set
```

```python
# import train_test_split class from the model_selection module of sklearn library
from sklearn.model_selection import train_test_split

# split the dataset into training and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
# view X_train
print(X_train)
```

```python
print(Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
# since the values in the training sets are very variable as compared to the test set, we will
# perform feature scaling on the training sets
```

```python
# 3. Feature Scaling
```

```python
# import StandardScaler class from preprocessing module of the sklearn library
from sklearn.preprocessing import StandardScaler

# create an instance for StandardScaler
standardscaler = StandardScaler()

# train the instance on the training set
X_train = standardscaler.fit_transform(X_train)

# train the instance on the testing set
X_test = standardscaler.transform(X_test)
```

```python
# view X_train
print(X_train)
```

```python
# view X_test
print(X_test)
```

```python
# 4. Train KNN classification model on the training set
```

```python
# import LogisticRegression class from the linear_model module of the sklearn library
from sklearn.neighbors import KNeighborsClassifier

# create a instance for KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)

# train the instance on the training sets
classifier.fit(X_train, Y_train)
```

```python
# 5. Predict a new result
```

```python
# try to predict if a customer of Age = 30 and Estimated Salary = 87000 will buy a product or not
print(classifier.predict(standardscaler.transform([[30,87000]])))
```

```python
# the model accurately predicts that the customer doesn't buy the product
```

```python
# 6. now we know that model is working, let's use the model to predict the test set results
```

```python
# predict Y values for X_test
Y_predict = classifier.predict(X_test)

# print predicted and test set
print(np.concatenate((Y_predict.reshape(len(Y_predict),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
# it seems that the model is making accurate predictions
```

```python
# 7. Make the confusion matrix
```

```python
# import confusion_matrix and accuracy_score from metrics module of sklearn library
from sklearn.metrics import confusion_matrix, accuracy_score

# create the confusion matrix
confusionmatrix = confusion_matrix(Y_test, Y_predict)

# print the confusionmatrix
print(confusionmatrix)

# show the accuracy score
accuracy_score(Y_test, Y_predict)
```

```python
# the logistic regression model seems to have an accuracy score of 93%, where 64 correct predictions for 0 (customers who didn't buy a SUV)
# and 29 correct predictions for 1 (customers who bought the SUV).
# Let's try to visualize the results now
```

```python
# 8. Visualization
```

```python
# view training set results
```

```python
from matplotlib.colors import ListedColormap

X_set, Y_set = standardscaler.inverse_transform(X_train), Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))

plt.contourf(X1, X2, classifier.predict(standardscaler.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(Y_set)):
    plt.scatter(X_set[Y_set == j, 0], X_set[Y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
    
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
from matplotlib.colors import ListedColormap

X_set, Y_set = standardscaler.inverse_transform(X_test), Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))

plt.contourf(X1, X2, classifier.predict(standardscaler.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(Y_set)):
    plt.scatter(X_set[Y_set == j, 0], X_set[Y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
    
plt.title('K-NN (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
